**상관계수(Correlation Coefficient)**
* 피어슨 상관계수: 두 변수 간 선형 관계의 강도
* 스피어만 상관계수: 순위 기반 상관관계
* 0 ~ 1 사이 값, 0에 가까우면 상관관계 약함

**주의사항**
* 상관관계 ≠ 인과관계
* 이상치에 민감함
* 비선형 관계는 포착 못할 수 있음

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Mac 한글 폰트 설정
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 샘플 데이터 생성
np.random.seed(42)
study_hours = np.random.uniform(1, 10, 100)
test_scores = 50 + 5 * study_hours + np.random.normal(0, 5, 100)

df = pd.DataFrame({
    'study_hours': study_hours,
    'test_scores': test_scores
})

# 상관계수 계산
correlation = df['study_hours'].corr(df['test_scores'])
print(f"상관계수: {correlation:.3f}")
```

```python
# 산점도와 상관관계 시각화
plt.figure(figsize=(10, 6))
plt.scatter(df['study_hours'], df['test_scores'], alpha=0.6)
plt.xlabel('Study Hours')
plt.ylabel('Test Scores')
plt.title(f'Correlation: {correlation:.3f}')
plt.show()

# 상관행렬 히트맵
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# 1. 여러 변수 간 상관행렬 만들고 해석하기

# 샘플 데이터 생성
np.random.seed(42)
exercise_hours = np.random.uniform(1, 6, 100)
muscle_mass = 50 + 5 * exercise_hours + np.random.normal(0, 5, 100)

df = pd.DataFrame({
    'exercise_hours': exercise_hours,
    'muscle_mass': muscle_mass
})

# 상관계수 계산
correlation = df['exercise_hours'].corr(df['muscle_mass'])
print(f"상관계수: {correlation:.3f}")

# 산점도와 상관관계 시각화
plt.figure(figsize=(10, 6))
plt.scatter(df['exercise_hours'], df['muscle_mass'], alpha=0.6)
plt.xlabel('운동 시간')
plt.ylabel('근육량')
plt.title(f'Correlation: {correlation:.3f}')
plt.show()

# 상관행렬 히트맵
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# 3. 이상치가 상관계수에 미치는 영향 실험하기

df = pd.DataFrame({
    'exercise_hours': exercise_hours,
    'muscle_mass': muscle_mass
})

# 상관계수 계산
correlation = df['exercise_hours'].corr(df['muscle_mass'])
print(f"상관계수: {correlation:.3f}")

# 이상치 투입
df.loc[4, 'muscle_mass'] = 10000

correlation = df['exercise_hours'].corr(df['muscle_mass'])
print(f"상관계수: {correlation:.3f}")
```
