### 확률

* 어떤 사건이 일어날 가능성을 0과 1사이의 숫자로 표현
* X라는 사건이 일어날 확률을 P(X)로 표현함

**표본공간**: 가능한 모든 결과의 집합
* 주사위: {1, 2, 3, 4, 5, 6}

**사건**: 표본 공간의 부분집합
* 짝수가 나오는 사건: {2, 4, 6}

**확률의 계산**
P(A) = 사건 A가 일어나는 경우의 수 / 전체 경우의 수

**확률의 기본 법칙**
* 덧셈 법칙: P(A or B): P(A) + P(B) - P(A and B)
* 곱셈 법칙: P(A and B): P(A) x P(B|A) ※A가 일어날 때 B가 일어날 확률

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'AppleGothic'
```

```python
# 1. 동전 던지기 시뮬레이션
np.random.seed(42) # 랜덤의 시드 값을 42로 설정
coin_flips = np.random.choice(['H', 'T'], size=1000) # H 또는 T를 무작위로 1000번 생성
print(f"앞면 확률: {np.mean(coin_flips == 'H'):.3f}")
```

```python
# 2. 주사위 던지기 시뮬레이션
dice_rolls = np.random.randint(1, 7, size=10000) # 1부터 6(포함)까지 10000개 뽑기
plt.hist(dice_rolls, bins=6, edgecolor='black') # 막대그래프(hist) 그리기, 간격 크기 6
plt.title('주사위 10,000번 던지기')
plt.xlabel('눈의 수')
plt.ylabel('빈도')
plt.show()
```

```python
# 2-1. 짝수가 나올 확률
even_probability = np.mean(dice_rolls % 2 == 0) # 짝수만 모아 평균 냄
print(f"짝수 확률: {even_probability:.3f}")
```

```python
# 3. 카드 뽑기 시뮬레이션
suits = ['♠', '♥', '♦', '♣'] # 트럼프 카드 모양은 4종
ranks = list(range(1, 14)) # 트럼프 카드 수는 1 ~ 13
deck = [(suit, rank) for suit in suits for rank in ranks] # 리스트 컴프리헨션 ('♠', 1) 꼴로 저장
```

```python
# 3-1. 52장에서 5장 뽑기
sample_hand = np.random.choice(len(deck), size=5, replace=False) # (덱)크기에서 5개의 수를 중복없이 추출
print(sample_hand) # 인덱스만 존재
hand = [deck[i] for i in sample_hand] # 그 번호에 해당하는 카드를 골라 리스트로 생성

print(f"뽑은 카드: {hand}")
```

```python
# 3-2. 패에서 한 장을 뽑았을 때 하트가 나올 확률
hearts_prob = len([c for c in hand if c[0] == '♥']) / len(hand)
print(f"하트 확률: {hearts_prob:.3f}")
```
