### 평균(Mean)
* 모든 값을 더해 개수로 나눔
* 이상치(outlier)에 민감

### 중앙값(Median)
* 데이터를 크기 순으로 정렬했을 때 가운데 값
* 이상치에 강함(robust)
* 데이터가 치우쳐 있을 때 평균보다 대표성이 좋음

### 최빈값(Mode)
* 가장 자주 나타나는 값
* 범주형 데이터에서 유용

#### 각 값은 언제 사용하는가?
* **평균값**: 데이터가 대칭적
* **중앙값**: 데이터가 치우침(skewed) 또는 이상치 존재
* **최빈값**: 범주형 데이터

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.font_manager as fm
import matplotlib as mpl

mpl.rcParams['font.family'] = 'AppleGothic'
mpl.rcParams['axes.unicode_minus'] = False
```

```python
# 1. 기본 통계량 계산
data = np.array([1, 2, 3, 4, 5, 5, 6, 7, 8, 9, 100])  # 100은 이상치

print("평균:", np.mean(data))
print("중앙값:", np.median(data))
print("최빈값:", pd.Series(data).mode()[0])
```

```python
# 이상치가 평균에 미치는 영향 시각화
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# 정상 데이터
normal_data = np.array([1, 2, 3, 4, 5, 5, 6, 7, 8, 9, 10])
axes[0].hist(normal_data, bins=10, edgecolor='black', alpha=0.7)
axes[0].axvline(np.mean(normal_data), color='r', linestyle='--', label=f'평균: {np.mean(normal_data):.1f}')
axes[0].axvline(np.median(normal_data), color='g', linestyle='--', label=f'중앙값: {np.median(normal_data):.1f}')
axes[0].set_title('정상 데이터')
axes[0].legend()

# 이상치 포함
data = np.array([1, 2, 3, 4, 5, 5, 6, 7, 8, 9, 100])  # 100은 이상치
axes[1].hist(data, bins=10, edgecolor='black', alpha=0.7)
axes[1].axvline(np.mean(data), color='r', linestyle='--', label=f'평균: {np.mean(data):.1f}')
axes[1].axvline(np.median(data), color='g', linestyle='--', label=f'중앙값: {np.median(data):.1f}')
axes[1].set_title('이상치 포함 데이터')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
# 2. 실제 데이터로 실습 - 소득 분포 시뮬레이션
np.random.seed(42)

# np.random.normal(mean, std, n): 정규분포에서 평균과 표준편차로 n개의 수 생성
# np.random.uniform(mean, std, n): 균등분포에서 평균과 표준편차로 n개의 수 생성

# 대부분의 사람들 소득 (정규분포)
normal_income = np.random.normal(50000, 15000, 950)
# 고소득자 (이상치)
high_income = np.random.uniform(200000, 1000000, 50)
income_data = np.concatenate([normal_income, high_income])

print(f"\n소득 데이터 분석:")
print(f"평균 소득: ${np.mean(income_data):,.0f}")
print(f"중앙값 소득: ${np.median(income_data):,.0f}")
print(f"최솟값: ${np.min(income_data):,.0f}")
print(f"최댓값: ${np.max(income_data):,.0f}")

print("\n-> 평균은 이상치에 민감하고 중앙값은 이상치에 강건하다")
```

```python
# 시각화
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(income_data, bins=50, edgecolor='black', alpha=0.7)
plt.axvline(np.mean(income_data), color='r', linestyle='--',
            label=f'평균: ${np.mean(income_data):,.0f}')
plt.axvline(np.median(income_data), color='g', linestyle='--',
            label=f'중앙값: ${np.median(income_data):,.0f}')
plt.xlabel('소득 ($)')
plt.ylabel('빈도')
plt.title('소득 분포 (전체)')
plt.legend()

plt.subplot(1, 2, 2)
plt.boxplot(income_data)
plt.ylabel('소득 ($)')
plt.title('소득 분포 (Box Plot)')

plt.tight_layout()
plt.show()
```

```python
# 3. Pandas로 그룹별 통계
df = pd.DataFrame({
    '부서': np.random.choice(['개발', '마케팅', '영업'], 100),
    '연봉': np.random.normal(60000, 20000, 100),
    '경력': np.random.randint(0, 20, 100)
})

print("\n부서별 평균 연봉:")
print(df.groupby('부서')['연봉'].mean())

print("\n부서별 중앙값 연봉:")
print(df.groupby('부서')['연봉'].median())

# describe()로 한 번에 보기
print("\n전체 통계 요약:")
print(df.describe())
```
