```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.font_manager as fm
import matplotlib as mpl

mpl.rcParams['font.family'] = 'AppleGothic'
mpl.rcParams['axes.unicode_minus'] = False
```

```python
# 데이터 생성 - 학생 성적 데이터
np.random.seed(42)

n_students = 200
df = pd.DataFrame({
    '학번': range(1, n_students + 1),
    '성별': np.random.choice(['남', '여'], n_students),
    '학년': np.random.choice([1, 2, 3, 4], n_students),
    '전공': np.random.choice(['컴공', '경영', '디자인'], n_students),
    '수학': np.random.normal(75, 15, n_students),
    '영어': np.random.normal(70, 20, n_students),
    '프로그래밍': np.random.normal(80, 12, n_students),
    '출석률': np.random.uniform(60, 100, n_students),
    '과제제출률': np.random.uniform(50, 100, n_students)
})
```

```python
# 1. 성별과 학년을 동시에 고려한 그룹별 평균 점수

df.groupby(['성별', '학년'])[['수학', '영어', '프로그래밍']].mean()
```

```python
# 2. 수학 점수가 90점 이상인 학생들의 특징을 분석

# 전공 가져오기
majors = df['전공'].unique()

# 90점 이상인 학생들 마스킹
idx = df['수학'] > 90
df_high = df[idx]

# 상관관계 컬럼들
correlation_cols = ['수학', '출석률', '과제제출률']

for major in majors:
    # 순차적으로 전공 1개만 마스킹
    sub = df_high[df_high['전공'] == major][correlation_cols]
    corr_matrix = sub.corr()

    plt.figure(figsize=(6, 5))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0,
                square=True, linewidths=1, cbar_kws={"shrink": 0.8})
    plt.title(f'{major} 전공 – 90점 이상 학생 상관관계', fontsize=14)
    plt.tight_layout()
    plt.show()
```

```python
# 3. 각 전공에서 수학 점수가 가장 높은 학생 3명 찾기
top3 = (
    df.sort_values(['전공', '수학'], ascending=[True, False])
      .groupby('전공')
      .head(3)
)

print(top3)
```
