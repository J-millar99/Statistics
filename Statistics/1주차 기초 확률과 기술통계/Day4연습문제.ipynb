```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.font_manager as fm
import matplotlib as mpl

mpl.rcParams['font.family'] = 'AppleGothic'
mpl.rcParams['axes.unicode_minus'] = False
```

```python
# 각 반의 시험 점수
A = np.array([85, 87, 86, 88, 84, 86, 85, 87, 86])
B = np.array([70, 95, 60, 100, 75, 90, 65, 95, 80])

# 반 전체 학생이므로 ddof=0
A_mean = np.mean(A); A_std = np.std(A, ddof=0)
B_mean = np.mean(B); B_std = np.std(B, ddof=0)

print(f"A반의 시험 점수 평균: {A_mean:.3f}", f"A반의 시험 점수 표준편차: {A_std:.3f}")
print(f"B반의 시험 점수 평균: {B_mean:.3f}", f"B반의 시험 점수 표준편차: {B_std:.3f}")
```

```python
df = pd.DataFrame({
    '그룹A': A,
    '그룹B': B
})

desc = df.describe()

# 반 전체가 모집단이기 때문
desc.loc['std'] = df.std(ddof=0)

print("\n그룹별 통계:")
print(desc)
```

```python
plt.figure(figsize=(10, 6))
df.boxplot()
plt.ylabel('점수')
plt.title('그룹별 점수 분포 비교')
plt.show()
```
