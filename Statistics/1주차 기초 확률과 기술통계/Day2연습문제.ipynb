### 연습문제 (20분)

1. 질병 검사의 민감도가 90%, 특이도가 99%이고 유병률이 5%일 때, 양성 판정을 받은 사람이 실제로 질병일 확률을 베이즈 정리로 계산하세요.
2. 위 문제를 시뮬레이션으로 검증하세요.

```python
# 유병률 5%
# 검사 민감도 90%: 질병 있을 때 양성
# 검사 특이도 99%: 질병 없을 때 음성
```

```python
import numpy as np
```

```python
# 1. 조건부확률로 계산
def simulate_medical_test1(n_people=100000):
    has_disease = np.random.random(n_people) < 0.05

    test_result = np.zeros(n_people, dtype=bool)

    # 검사 민감도 90%: 양성판정의 90%만 믿을 수 있음
    test_result[has_disease] = np.random.random(has_disease.sum()) < 0.90

    # 검사 특이도 99%: 음성판정의 99%만 믿을 수 있음
    test_result[~has_disease] = np.random.random((~has_disease).sum()) < 0.01

    # 양성 판정을 받은 사람이 실제 질병을 가질 확률: 질병환자 수 / 양성 판정 수
    positive_test_rate = has_disease[test_result].sum() / test_result.sum()
    print(f"\n양성 판정을 받은 사람이 실제로 질병일 확률: {positive_test_rate:.3f}")

simulate_medical_test1()
```

```python
# 2. 베이즈 정리로 계산

def simulate_mdeical_test2(n_peolple=100000):
    # 유병률 5%
    has_disease = 0.05

    # 검사 민감도 99% P(양성|질병)
    positive_with_disease = 0.90

    # 검사 특이도 99% P(음성|정상)
    negative_with_nodisease = 0.99

    # 1% P(양성\정상)
    positive_with_nodisease = 0.01

    positive = positive_with_disease * has_disease + positive_with_nodisease * (1 - has_disease)
    # 질병|양성 = 양성|질병 * 질병 / 양성
    disease_with_positive = positive_with_disease * has_disease / positive

    print(f"\n양성 판정을 받은 사람이 실제로 질병일 확률: {disease_with_positive:.3f}")

simulate_mdeical_test2()
    
```
