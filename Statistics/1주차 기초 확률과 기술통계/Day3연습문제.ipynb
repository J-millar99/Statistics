```python
import numpy as np
import pandas as pd
```

```python
scores = np.array([85, 90, 75, 90, 95, 80, 90, 70, 85, 100])
```

```python
print(f"시험 점수 데이터의 평균값: {np.mean(scores):.2f}")
print(f"시험 점수 데이터의 중앙값: {np.median(scores):.2f}")
print(f"시험 점수 데이터의 최빈값: {pd.Series(scores).mode()[0]}")
```

```python
# 한 명의 학생이 0점을 받은 경우 각 값이 어떻게 변할까?

# 기존 점수 배열
print(scores)
print(f"시험 점수 데이터의 평균값: {np.mean(scores):.2f}")
print(f"시험 점수 데이터의 중앙값: {np.median(scores):.2f}")

# 특정 배열 값을 0으로 교체
scores[3] = 0
print(scores)

# 변화 비교
print(f"시험 점수 데이터의 평균값: {np.mean(scores):.2f}")
print(f"시험 점수 데이터의 중앙값: {np.median(scores):.2f}")
```
