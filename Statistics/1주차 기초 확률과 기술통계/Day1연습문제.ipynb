### 연습문제

1. 두 개의 주사위를 던질 때 합이 7이 나올 확률을 시뮬레이션으로 구하세요.
2. 52장 카드에서 5장을 뽑을 때 모두 같은 무늬일 확률을 10,000번 시뮬레이션으로 추정하세요.

```python
# 1. 두 개의 주사위를 던질 때 합이 7이 나올 확률을 시뮬레이션으로 구하세요.
import numpy as np

tries = 10000
count = 0
for _ in range(tries):
    dice_rolls = np.random.randint(1, 7, size=2)
    if sum(dice_rolls) == 7:
        count += 1

print(f"두 주사위의 합이 7이 나올 확률 {(count/tries):.3f}")
```

```python
# 2. 52장 카드에서 5장을 뽑을 때 모두 같은 무늬일 확률을 10,000번 시뮬레이션으로 추정하세요.

suits = ['♠', '♥', '♦', '♣'] # 트럼프 카드 모양은 4종
ranks = list(range(1, 14)) # 트럼프 카드 수는 1 ~ 13
decks = [(suit, rank) for suit in suits for rank in ranks]

tries = 10000
count = 0

for _ in range(tries):
    sample_index = np.random.choice(len(decks), size=5, replace=False)
    hand = [decks[idx] for idx in sample_index]
    if all(x[0] == hand[0][0] for x in hand):
        count += 1

print(f"뽑은 5장의 카드 무늬가 모두 같을 확률: {(count/tries):.3f}")
```
