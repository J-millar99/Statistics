### 포아송 분포
* 단위 시간/공간에서 사건이 일어나는 횟수
  * 1시간 동안 웹 사이트 방문자수
  * 하루 동안 고객 문의 건수
  * $1km^2$당 사고 발생 건수
* 드문 사건, 독립적으로 사건이 발생
* 매개변수 λ: 평균 발생 횟수를 의미
* PMF: P(X=k) = $(λ^k × e^{-λ}) / k!$

### 이항분포 vs 포아송 분포
* 이항분포: 시행 횟수 n이 고정, 성공 확률 p 작음
* 포아송 분포: n이 매우 크고 p가 매우 작을 때 근사 (np = λ)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

import matplotlib as mpl

# 한글 폰트 설정 (선택사항)
mpl.rcParams['font.family'] = 'AppleGothic'
mpl.rcParams['axes.unicode_minus'] = False
```

```python
# 1. 포아송 분포 기본
print("=== 포아송 분포 ===")
# 시간당 평균 고객 문의 3건
lambda_val = 3
poisson_dist = stats.poisson(lambda_val)

# 확률 계산
print(f"문의 0건일 확률: {poisson_dist.pmf(0):.4f}")
print(f"문의 3건일 확률: {poisson_dist.pmf(3):.4f}")
print(f"문의 5건 이상일 확률: {1 - poisson_dist.cdf(4):.4f}")
```

```python
# 2. 다양한 λ 값에 따른 분포 변화
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
lambdas = [1, 3, 5, 10]

for idx, lam in enumerate(lambdas):
    ax = axes[idx // 2, idx % 2]

    dist = stats.poisson(lam)
    x = np.arange(0, lam*3)
    pmf = dist.pmf(x)

    ax.bar(x, pmf, alpha=0.7, color='teal', edgecolor='black')
    ax.axvline(lam, color='red', linestyle='--', linewidth=2, label=f'λ = {lam}')
    ax.set_xlabel('발생 횟수 (k)')
    ax.set_ylabel('확률')
    ax.set_title(f'포아송 분포 (λ = {lam})')
    ax.legend()
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# 3. 실전 예제: 웹사이트 트래픽 분석
print("\n=== 실전: 웹사이트 트래픽 ===")
# 평균 시간당 방문자 50명
avg_visitors = 50

# 하루(24시간) 동안 시뮬레이션
hourly_visitors = stats.poisson(avg_visitors).rvs(size=24)

plt.figure(figsize=(14, 5))

# 왼쪽: 시간대별 방문자
plt.subplot(1, 2, 1)
hours = range(24)
plt.bar(hours, hourly_visitors, alpha=0.7, color='coral', edgecolor='black')
plt.axhline(avg_visitors, color='red', linestyle='--', label=f'평균 = {avg_visitors}')
plt.xlabel('시간 (0~23시)')
plt.ylabel('방문자 수')
plt.title('24시간 동안 시간대별 방문자')
plt.legend()
plt.grid(alpha=0.3)

# 오른쪽: 분포 히스토그램
plt.subplot(1, 2, 2)
plt.hist(hourly_visitors, bins=15, alpha=0.7, color='lightblue', edgecolor='black', density=True)
x_range = np.arange(30, 70)
plt.plot(x_range, stats.poisson(avg_visitors).pmf(x_range), 'r-', linewidth=2, label='이론적 분포')
plt.xlabel('시간당 방문자 수')
plt.ylabel('확률 밀도')
plt.title('방문자 분포')
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print(f"총 방문자: {sum(hourly_visitors)}명")
print(f"최대 방문 시간대: {np.argmax(hourly_visitors)}시 ({max(hourly_visitors)}명)")
print(f"최소 방문 시간대: {np.argmin(hourly_visitors)}시 ({min(hourly_visitors)}명)")
```

```python
# 4. 이항분포와 포아송 근사 비교
print("\n=== 이항분포 vs 포아송 근사 ===")
n = 1000
p = 0.003
lam = n * p  # λ = 3

x = np.arange(0, 15)
binom_pmf = stats.binom(n, p).pmf(x)
poisson_pmf = stats.poisson(lam).pmf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, binom_pmf, 'o-', label=f'이항분포 (n={n}, p={p})', markersize=8)
plt.plot(x, poisson_pmf, 's--', label=f'포아송 근사 (λ={lam})', markersize=8)
plt.xlabel('k')
plt.ylabel('확률')
plt.title('이항분포와 포아송 근사 비교')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
