### 표준정규분포
* 정의: 평균 0, 표준편차 1인 정규분포
* 기호: Z ~ N(0, 1)
* 모든 정규분포를 표준화할 수 있음

### Z-score
* 공식: Z = (X - μ) / σ
* 의미: "평균에서 표준편차 몇 개만큼 떨어져 있는가?"
* 활용: 서로 다른 척도를 비교할 수 있음

### Z-table
* 표준정규분포의 누적활률표
* 실무에서는 Python으로 계산

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

import matplotlib as mpl

# 한글 폰트 설정 (선택사항)
mpl.rcParams['font.family'] = 'AppleGothic'
mpl.rcParams['axes.unicode_minus'] = False
```

```python
# 1. 표준정규분포
print("=== 표준정규분포 ===")
z_dist = stats.norm(0, 1)

# 주요 확률값
print(f"P(Z ≤ 1.96) = {z_dist.cdf(1.96):.4f}")  # 95%
print(f"P(Z ≤ -1.96) = {z_dist.cdf(-1.96):.4f}")  # 2.5%
print(f"P(-1.96 ≤ Z ≤ 1.96) = {z_dist.cdf(1.96) - z_dist.cdf(-1.96):.4f}")  # 95%
```

```python
# 2. 표준정규분포 시각화
z = np.linspace(-4, 4, 1000)
pdf = z_dist.pdf(z)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(z, pdf, linewidth=2, color='darkblue')
plt.fill_between(z, pdf, where=(z>=-1.96) & (z<=1.96), alpha=0.3, color='red', label='95% 구간')
plt.axvline(0, color='black', linestyle='-', linewidth=1)
plt.axvline(-1.96, color='red', linestyle='--', linewidth=2, label='Z = -1.96')
plt.axvline(1.96, color='red', linestyle='--', linewidth=2, label='Z = 1.96')
plt.xlabel('Z')
plt.ylabel('확률 밀도')
plt.title('표준정규분포 N(0,1)')
plt.legend()
plt.grid(alpha=0.3)

# CDF
plt.subplot(1, 2, 2)
cdf = z_dist.cdf(z)
plt.plot(z, cdf, linewidth=2, color='green')
plt.axhline(0.5, color='black', linestyle='--', alpha=0.5)
plt.axhline(0.025, color='red', linestyle='--', alpha=0.5, label='2.5%')
plt.axhline(0.975, color='red', linestyle='--', alpha=0.5, label='97.5%')
plt.xlabel('Z')
plt.ylabel('누적 확률')
plt.title('표준정규분포 CDF')
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# 3. Z-score 계산 및 표준화
print("\n=== Z-score 예제 ===")

# 예제: 두 학생의 성적 비교
# 철수: 수학 85점 (평균 70, 표준편차 10)
# 영희: 영어 90점 (평균 80, 표준편차 15)

math_score = 85
math_mean, math_std = 70, 10
z_math = (math_score - math_mean) / math_std

english_score = 90
english_mean, english_std = 80, 15
z_english = (english_score - english_mean) / english_std

print(f"철수 수학 Z-score: {z_math:.2f}")
print(f"영희 영어 Z-score: {z_english:.2f}")
print(f"\n상대적으로 더 잘한 사람: {'철수' if z_math > z_english else '영희'}")

# 백분위 계산
percentile_math = z_dist.cdf(z_math) * 100
percentile_english = z_dist.cdf(z_english) * 100

print(f"\n철수는 상위 {100-percentile_math:.1f}%")
print(f"영희는 상위 {100-percentile_english:.1f}%")
```

```python
# 4. 실전 예제: 여러 과목 성적 표준화
print("\n=== 실전: 성적 표준화 ===")

# 5명의 학생, 3과목 성적
students = ['학생A', '학생B', '학생C', '학생D', '학생E']
math_scores = np.array([85, 92, 78, 88, 95])
english_scores = np.array([70, 85, 90, 75, 80])
science_scores = np.array([88, 82, 95, 90, 85])

# 표준화
z_math = (math_scores - math_scores.mean()) / math_scores.std()
z_english = (english_scores - english_scores.mean()) / english_scores.std()
z_science = (science_scores - science_scores.mean()) / science_scores.std()

# 종합 점수 (표준화 후 평균)
composite_z = (z_math + z_english + z_science) / 3

# 시각화
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 원점수
ax1 = axes[0, 0]
x = np.arange(len(students))
width = 0.25
ax1.bar(x - width, math_scores, width, label='수학', alpha=0.8)
ax1.bar(x, english_scores, width, label='영어', alpha=0.8)
ax1.bar(x + width, science_scores, width, label='과학', alpha=0.8)
ax1.set_xlabel('학생')
ax1.set_ylabel('점수')
ax1.set_title('과목별 원점수')
ax1.set_xticks(x)
ax1.set_xticklabels(students)
ax1.legend()
ax1.grid(alpha=0.3)

# Z-score
ax2 = axes[0, 1]
ax2.bar(x - width, z_math, width, label='수학', alpha=0.8)
ax2.bar(x, z_english, width, label='영어', alpha=0.8)
ax2.bar(x + width, z_science, width, label='과학', alpha=0.8)
ax2.axhline(0, color='black', linestyle='-', linewidth=1)
ax2.set_xlabel('학생')
ax2.set_ylabel('Z-score')
ax2.set_title('과목별 Z-score (표준화)')
ax2.set_xticks(x)
ax2.set_xticklabels(students)
ax2.legend()
ax2.grid(alpha=0.3)

# 종합 순위
ax3 = axes[1, 0]
colors = ['gold', 'silver', 'brown', 'lightblue', 'lightgreen']
sorted_indices = np.argsort(composite_z)[::-1]
ax3.barh(range(len(students)), composite_z[sorted_indices], color=[colors[i] for i in range(len(students))])
ax3.set_yticks(range(len(students)))
ax3.set_yticklabels([students[i] for i in sorted_indices])
ax3.set_xlabel('종합 Z-score')
ax3.set_title('종합 순위 (3과목 평균 Z-score)')
ax3.grid(alpha=0.3)

# 성적 분포 비교
ax4 = axes[1, 1]
ax4.boxplot([math_scores, english_scores, science_scores], labels=['수학', '영어', '과학'])
ax4.set_ylabel('점수')
ax4.set_title('과목별 점수 분포')
ax4.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# 결과 출력
print("\n=== 종합 순위 ===")
for rank, idx in enumerate(sorted_indices, 1):
    print(f"{rank}위: {students[idx]} (Z-score: {composite_z[idx]:.2f})")
```

```python
# 1. 키가 평균 165cm, 표준편차 8cm인 정규분포를 따릅니다. 175cm는 상위 몇 %인가요?
height_mean = 165
height_std = 8

dist = stats.norm(height_mean, height_std)

a = dist.cdf(175)

# 백분위 계산
h = dist.cdf(175) * 100

print(f"175cm는 상위 {100-h:.1f}%")
```

```python
# 2. Z-score가 -1.5인 값은 전체의 하위 몇 %에 해당하나요?
z_dist = stats.norm(0, 1)

rank = z_dist.cdf(-1.5) * 100
print(rank)
```

```python
# 3. 100명의 학생 성적 데이터를 생성하고(평균 75, 표준편차 15),
# 각 학생의 Z-score를 계산한 후 상위10%에 해당하는 학생들을 찾아보세요.

score_mean = 75
score_std = 15

# 1. 점수 생성
scores = np.random.normal(score_mean, score_std, 100)

# 2. Z-score 계산
z_scores = (scores - score_mean) / score_std

# 3. 상위 10% Z-score 기준 (90th percentile)
z_cut = stats.norm.ppf(0.90)

print(f"상위 10% Z-score 기준값: {z_cut:.4f}\n")

# 4. 상위 10% 학생 찾기
top10_idx = np.where(z_scores >= z_cut)[0]   # 인덱스(0~99)

top10_scores = scores[top10_idx]

# 결과 출력
for i, (idx, score) in enumerate(zip(top10_idx, top10_scores), 1):
    print(f"{i}. {idx+1}번 학생 — 점수: {score:.2f}, Z-score: {z_scores[idx]:.2f}")
```
