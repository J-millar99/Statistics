### 비모수 검정(카이제곱 검정)
* t-검정은 데이터가 정규분포를 따른다고 가정
* 범주형 데이터나 정규성을 만족하지 않을때는 비모수 검정 사용

### 카이제곱 검정
범주형 변수 간의 관계를 검정

**1. 적합도 검정**

관측된 빈도가 기대 빈도와 일치하는지 검정
* 예: 주사위가 공정한가?

**2. 독립성 검정**

두 범주형 변수가 독립인지 검정
* 예: 성별과 제품 선호도가 관련 있는가?

**카이제곱 통계량 계산**
* χ² = Σ [(관측빈도 - 기대빈도)² / 기대빈도]

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

# Mac 한글 폰트 설정
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
print("="*60)
print("1. 카이제곱 적합도 검정 (Goodness of Fit Test)")
print("="*60)
print("문제: 주사위가 공정한가?")
print("(각 면이 나올 확률이 모두 1/6인가?)")
print()

# 주사위를 60번 던진 결과
np.random.seed(42)
# 실제로는 약간 불공정한 주사위 (1이 많이 나옴)
dice_rolls = np.random.choice([1, 2, 3, 4, 5, 6], size=60, p=[0.25, 0.15, 0.15, 0.15, 0.15, 0.15])

# 관측 빈도
observed = np.bincount(dice_rolls, minlength=7)[1:]  # 0 제외
expected = np.array([10, 10, 10, 10, 10, 10])  # 각 면이 10번씩 나와야 함

print(f"귀무가설(H₀): 주사위는 공정하다 (각 면 확률 = 1/6)")
print(f"대립가설(H₁): 주사위는 공정하지 않다")
print()
print("관측 빈도:", observed)
print("기대 빈도:", expected)

# 카이제곱 검정
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

print(f"\n=== 검정 결과 ===")
print(f"카이제곱 통계량(χ²): {chi2_stat:.4f}")
print(f"자유도(df): {len(observed) - 1}")
print(f"p-value: {p_value:.4f}")
print(f"\n결론 (α=0.05): ", end="")
if p_value < 0.05:
    print("귀무가설 기각! 주사위가 공정하지 않습니다.")
else:
    print("귀무가설 채택. 주사위가 공정하다고 볼 수 있습니다.")
```

```python
print("\n" + "="*60)
print("2. 카이제곱 독립성 검정 (Test of Independence)")
print("="*60)
print("문제: 성별과 제품 선호도가 관련이 있는가?")
print()

# 교차표 생성
data = pd.DataFrame({
    '성별': ['남성']*60 + ['여성']*60,
    '선호제품': 
        ['A']*25 + ['B']*20 + ['C']*15 +  # 남성
        ['A']*15 + ['B']*30 + ['C']*15     # 여성
})

# 교차표
contingency_table = pd.crosstab(data['성별'], data['선호제품'])
print("교차표 (관측 빈도):")
print(contingency_table)
print()

print(f"귀무가설(H₀): 성별과 제품 선호도는 독립이다")
print(f"대립가설(H₁): 성별과 제품 선호도는 관련이 있다")

# 카이제곱 독립성 검정
chi2, p_value2, dof, expected_freq = stats.chi2_contingency(contingency_table)

print(f"\n기대 빈도:")
expected_df = pd.DataFrame(
    expected_freq, 
    index=contingency_table.index,
    columns=contingency_table.columns
)
print(expected_df)

print(f"\n=== 검정 결과 ===")
print(f"카이제곱 통계량(χ²): {chi2:.4f}")
print(f"자유도(df): {dof}")
print(f"p-value: {p_value2:.4f}")
print(f"\n결론 (α=0.05): ", end="")
if p_value2 < 0.05:
    print("귀무가설 기각! 성별과 제품 선호도는 관련이 있습니다.")
else:
    print("귀무가설 채택. 성별과 제품 선호도는 독립입니다.")
```

```python
# 시각화
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# 1) 주사위 적합도 검정
x = np.arange(1, 7)
width = 0.35
axes[0].bar(x - width/2, observed, width, label='관측 빈도', color='skyblue', edgecolor='black')
axes[0].bar(x + width/2, expected, width, label='기대 빈도', color='lightcoral', edgecolor='black')
axes[0].set_xlabel('주사위 면')
axes[0].set_ylabel('빈도')
axes[0].set_title('적합도 검정: 주사위 공정성\n(p={:.4f})'.format(p_value), fontweight='bold')
axes[0].set_xticks(x)
axes[0].legend()
axes[0].grid(True, alpha=0.3, axis='y')

# 2) 교차표 히트맵
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='Blues', ax=axes[1], 
            cbar_kws={'label': '빈도'})
axes[1].set_title('독립성 검정: 교차표\n(p={:.4f})'.format(p_value2), fontweight='bold')
axes[1].set_xlabel('선호 제품')
axes[1].set_ylabel('성별')

# 3) 그룹별 비교
contingency_pct = contingency_table.div(contingency_table.sum(axis=1), axis=0) * 100
contingency_pct.plot(kind='bar', ax=axes[2], color=['skyblue', 'lightgreen', 'lightcoral'], 
                     edgecolor='black', width=0.7)
axes[2].set_title('성별별 제품 선호도 비율', fontweight='bold')
axes[2].set_xlabel('성별')
axes[2].set_ylabel('비율 (%)')
axes[2].legend(title='제품', bbox_to_anchor=(1.05, 1))
axes[2].set_xticklabels(axes[2].get_xticklabels(), rotation=0)
axes[2].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

```python
# 추가 예제: 혈액형과 질병
print("\n" + "="*60)
print("3. 추가 예제: 혈액형과 질병 발생")
print("="*60)

blood_disease = pd.DataFrame({
    '혈액형': ['A', 'B', 'AB', 'O'],
    '질병있음': [30, 25, 10, 35],
    '질병없음': [70, 75, 40, 85]
})
blood_disease = blood_disease.set_index('혈액형')

print(blood_disease)
print()

chi2_3, p_3, dof_3, expected_3 = stats.chi2_contingency(blood_disease)

print(f"카이제곱 통계량: {chi2_3:.4f}")
print(f"p-value: {p_3:.4f}")
print(f"\n결론: ", end="")
if p_3 < 0.05:
    print("혈액형과 질병 발생은 관련이 있습니다.")
else:
    print("혈액형과 질병 발생은 독립입니다.")
```

```python
# 1. 주사위를 600번 던지면 검정 결과가 달라지나요?

np.random.seed(42)
# 실제로는 약간 불공정한 주사위 (1이 많이 나옴)
dice_rolls = np.random.choice([1, 2, 3, 4, 5, 6], size=600, p=[0.25, 0.15, 0.15, 0.15, 0.15, 0.15])

# 관측 빈도
observed = np.bincount(dice_rolls, minlength=7)[1:]  # 0 제외
expected = np.array([100, 100, 100, 100, 100, 100])  # 각 면이 100번씩 나와야 함

# 카이제곱 검정
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

print(f"\n=== 검정 결과 ===")
print(f"p-value: {p_value:.4f}")

if p_value < 0.05:
    print("귀무가설 기각! 주사위가 공정하지 않습니다.")
else:
    print("귀무가설 채택. 주사위가 공정하다고 볼 수 있습니다.")

# 더 많이 1이 나올것이므로 불공정
```

```python
# 2. 성별과 학과 선택의 독립성을 검정하는 데이터를 만들어보세요.

print("\n" + "="*60)
print("2. 카이제곱 독립성 검정 (Test of Independence)")
print("="*60)
print("문제: 성별과 학과 선호도가 관련이 있는가?")
print()

# 교차표 생성
data = pd.DataFrame({
    '성별': ['남성']*60 + ['여성']*60,
    '선호학과': 
        ['컴과']*25 + ['생명']*20 + ['사회']*15 +  # 남성
        ['컴과']*15 + ['생명']*30 + ['사회']*15     # 여성
})

# 교차표
contingency_table = pd.crosstab(data['성별'], data['선호학과'])
print("교차표 (관측 빈도):")
print(contingency_table)
print()

print(f"귀무가설(H₀): 성별과 학과 선호도는 독립이다")
print(f"대립가설(H₁): 성별과 학과 선호도는 관련이 있다")

# 카이제곱 독립성 검정
chi2, p_value2, dof, expected_freq = stats.chi2_contingency(contingency_table)

print(f"\n기대 빈도:")
expected_df = pd.DataFrame(
    expected_freq, 
    index=contingency_table.index,
    columns=contingency_table.columns
)
print(expected_df)

print(f"\n=== 검정 결과 ===")
print(f"카이제곱 통계량(χ²): {chi2:.4f}")
print(f"자유도(df): {dof}")
print(f"p-value: {p_value2:.4f}")
print(f"\n결론 (α=0.05): ", end="")
if p_value2 < 0.05:
    print("귀무가설 기각! 성별과 학과 선호도는 관련이 있습니다.")
else:
    print("귀무가설 채택. 성별과 학과 선호도는 독립입니다.")
```
